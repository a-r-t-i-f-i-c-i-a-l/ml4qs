```python
by_subject = smartphone_train.groupby('subject')
n_subjects = by_subject.ngroups

fig, ax = plt.subplots(n_subjects, figsize=(10, 15))
fig.suptitle('Comparison of mean acceleration for walking in X, Y, Z axes across subjects')
i = 0
for name, group in by_subject:
  ax[i].plot(group[group.Activity == 'WALKING']['tBodyAcc-mean()-X'].reset_index(drop=True))
  ax[i].plot(group[group.Activity == 'WALKING']['tBodyAcc-mean()-Y'].reset_index(drop=True))
  ax[i].plot(group[group.Activity == 'WALKING']['tBodyAcc-mean()-Z'].reset_index(drop=True))
  i += 1

fig.tight_layout()
fig.subplots_adjust(top=0.95)

```
